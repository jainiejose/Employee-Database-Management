# SQL scripts for the online retail project




(1) SQL technique #1 - Counting rows, select and distinct functions

Count Function: Query the information to see how many rows are there:

```python
SELECT COUNT (*)
FROM Online_retail;
```

There is a total of 525,461 rows in this dataset

Select Function: This function extracts data from the enter database

```python
SELECT *
FROM Online_retail;
```

Sometimes, the dataset might be large and instead of querying the entire dataset you want to query a few lines to view what is in the table. In such scenarious we can use the LIMIT function:

```python
SELECT *
FROM Online_retail
LIMIT 10;
```

This will display the first 10 rows of the table

Distinct Function: A lot of times, our data table is filled with duplicate values. To attain the unique value, we use the DISTINCT function. This function will be especially helpful when working with large datasets.
            


In our dataset, how can we find the unique countries where the customers are residing?

```python
Select DISTINCT country
FROM Online_retail;
```

There are a total of 40 different countries


(2) SQL technique #2 - Aggregation functions

Aggregation functions are the base of any kind of data analysis. They provide us with an overview of the dataset. Some of the functions I will be querying are – MIN(), MAX(), SUM() and AVG().

The MIN() function returns the smallest value of the selected column.

```python
SELECT Country,
MIN(price) as MIN_price
FROM Online_retail
GROUP BY Country;
```

In this example, we are finding the min product price per unit for each country.

Calculate the SUM: The SUM() function is to calculate the sum of the numerical column in a table. Let’s find out the sum of the total of quantity of all the "PINK FLORAL FELTCRAFT SHOULDER BAGS"

```python
SELECT Description,
SUM(quantity) AS sum_quantity
FROM Online_retail
WHERE Description= 'PINK FLORAL FELTCRAFT SHOULDER BAG';
```

Calculate the AVERAGE: To calculate the average of the numeric columns, we use the AVG() function. Let’s find the average unit price:

```python
SELECT AVG(price) as Average_price
FROM Online_retail;
```

The average amount spent by customers in the online retail store is 4.68 sterlings


(3) SQL Technique #3 - Slicing Data


In this technique we will be going to query and retrieve data based on some kind of condition. 

For example, Let's say that the retail store wants to find the invoice number and the price of the items for those customers who live in Finland

```python
SELECT invoice, country, price
FROM (SELECT * FROM Online_retail WHERE country ='Finland');
```

Let's check out another example to slice our data.

This time the retail store wants to retrieve all the customer ID & invoice number of those customers who are spending between 15 and 20 sterlings so as to push out special marketing efforts.


```python
SELECT "Customer ID", Invoice, Price 
From Online_retail
WHERE Price BETWEEN 15 AND 20;
```

(4) SQL technique #4 - Sorting Data

Sorting data helps us put our data into perspective. We can perform the sorting porcess by using the - Order By

Let us see an example where we sort the data according to the column Price in ascending order where price is greater than 0:


```python
SELECT *
FROM Online_retail
WHERE Price >0
ORDER BY Price;
```

(5) SQL Technique #5 - Filtering Patterns

In this last section we will work on how to filter the columns that match a specific pattern. Let's query the description of all the items that ends with "bracelet" in their respective descriptions 

```python
SELECT *
FROM Online_retail
WHERE Description LIKE "%bracelet";
```
